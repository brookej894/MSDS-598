```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import random

import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

### Question 1 
Suppose we want to use the variables distance and passengers to predict total which represents the total fare. 
#### A. Use linear regression to create a model prediction total from distance and passengers. Report the ${R^2}$


```python
model = ols(formula = 'total ~ distance + passengers', data=df_taxis)
res = model.fit()
res.summary()
```

#### Answer:
${R^2}$ = 0.818

#### B. Incorporate another variable taxi color into the model and then report the ${R^2}$.

```python
model2 = ols(formula = 'total ~ distance + passengers + color', data=df_taxis)
res2 = model2.fit()
res2.summary()
```

#### Answer:
${R^2}$ = 0.829

By incorporating taxi color into the regression, the ${R^2}$ value increased, indicating that our model does a better job at explaining the variation in total fare when we include the taxi color. Logically, this doesn't seem like it makes a lot of sense. If a researcher were approaching this problem with background research, I doubt they would theorize that the car color impacts the total fare. 

### Question 2
As in the fourth lecture, create a length of ride variable. Use the new feature you've engineered to further improve the model and report the ${R^2}$.

```python
delta = pd.to_datetime(df_taxis['dropoff']) - pd.to_datetime(df_taxis['pickup'])
df_taxis['length_of_trip'] = delta / pd.Timedelta('60s')
```

```python
model3 = ols(formula = 'total ~ distance + passengers + color + length_of_trip', data=df_taxis)
res3 = model3.fit()
res3.summary()
```

#### Answer:
${R^2}$ = 0.847

By including the length of the trip we improve the explanatory power of our model but risk over-fitting by including more variables than we need.

### Question 3
Create a valifation set using ten percent of the data, and use the complement to create a training set. (Make sure there is no overlap in observation between the validation and training set).

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df_taxis, test_size=0.1)
```

### Question 4
Use this validation and the ${R^2}$ metric to decide whether adding length of ride to the variables from Q1 improves the model.

```python
# build model on training set
model4 = ols(formula = 'total ~ distance + passengers + color + length_of_trip', data=train)
res4 = model4.fit()

# predict using the model
y_pred = res4.predict(test)

# score the model
y = test['total']
y_mean = train['total'].mean()  

TSS = sum((y_pred - y_mean)**2)
RSS = sum((y_pred - y)**2)
print(f'This model has an R^2 on the test set of {(TSS - RSS) / TSS}')
```

#### Answer:
${R^2}$ = 0.8605

By including more variables and using validation to test our model it seems that the explanatory power is improving since the ${R^2}$ is increasing. However, we can logically surmise that some of these variables may be adding extraneous value to the ${R^2}$ metric without telling us about useful relationships, such as car color. As previously mentioned, car color logically shouldn't affect the total fare. Part of the increase in the measure of explanatory power therefore might be coming from explaining more variance that doesn't help explain the important relationships. 

### Question 5
Engineer one new feature from the data and repeat Q4. (Hint: you can use powers, differences, ratios, products and other features)

```python
df_taxis.head()
```

```python
# New Feature: average speed per trip
df_taxis['trip_length_hours'] = df_taxis['length_of_trip']/60
df_taxis['average_speed'] = df_taxis['distance']/df_taxis['trip_length_hours']
df_taxis.head()
```

```python
train, test = train_test_split(df_taxis, test_size=0.1)

# build model on training set
model5 = ols(formula = 'total ~ distance + passengers + color + average_speed', data=train)
res5 = model5.fit()

# predict using the model
y_pred = res5.predict(test)

# score the model
y = test['total']
y_mean = train['total'].mean()  

TSS = sum((y_pred - y_mean)**2)
RSS = sum((y_pred - y)**2)
print(f'This model has an R^2 on the test set of {(TSS - RSS) / TSS}')
```

#### Answer:
${R^2}$ = 0.911

The ${R^2}$ improved quite a lot, and the model looks like it has very high explanatory power from adding in the average speed of each trip.
